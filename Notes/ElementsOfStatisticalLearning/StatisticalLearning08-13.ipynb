### Simple linear regression

We assume a model:

$$ Y = \beta_{0} + \beta_{1} X + \epsilon $$

where $\beta_{0}, \beta_{1}$ are two unknown constants that represent the *intercept* and *slope*, aka *coefficients* or *parameters*, and $\epsilon$ is the error term.

### Estimations of parameters

Let $\hat{y}_{i} = \hat{\beta}_{0} + \hat{\beta}_{1}x_{i}$ be the prediction for $Y$ based on the $i$th value of X. Then $e_{i} = y_{i} - \hat{y}_{i}$ represents the $i$th *residual*

* residual looks like the vertical distance from the point to the line.

We choose $\hat{\beta}_{1}$ and $\hat{\beta}_{0}$ to be:

$$ \hat{\beta}_{1} = \dfrac {\sum_{i=1}^{n} (x_{i} - \bar{x})(y_{i} - \bar{y})}{\sum_{i=1}^{n} (x_{i} - \bar{x})^2}, $$ 

$$ \beta_{0} = \bar{y} - \hat{\beta}_{1}\bar{x}, $$

where $\bar{y} = \frac {1}{n} \sum_{i = 1} ^ {n} y_{i} $ and $\bar{x} = \frac {1}{n} \sum_{i = 1} ^ {n} x_{i} $ are the sample means.

### Assessing the Accuracy

We looks at the standard error of an estimator:

$$ SE(\hat{\beta}_{1})^2 = \dfrac{\sigma ^ 2}{\sum_{i = 1} ^ {n} (x_{i} - \bar{x}) ^ 2}, SE(\hat{\beta}_{0})^2 = \sigma ^ 2[\dfrac{1}{n} + \dfrac{\bar{x} ^ 2} {\sum_{i = 1} ^ {n} (x_{i} - \bar{x}) ^ 2}]$$ 

where $\sigma ^ 2 = Var(\epsilon) $

* $\sigma ^ 2$ is the "noise"
* the more spread out $x$ is, smaller $SE(\beta_{1})$ is
* these SEs can be used to compute *confidence intervals*. A 95% confidence interval is defined as a range of values such that with 95% probability, the range will contain the true unknown value of the parameter. It has the form:

$$\hat{\beta} \pm 2 * SE(\hat{\beta}_{1}) $$

i.e. there's a 95% change that the above interval will contain the true value of $\beta_{1}$.

** p-value is %-change that this data is significant**

* the CI for TV was [0.042, 0.053]

## Hypothesis Testing (09)

SEs can be used to perform hypothesis tests on the coefficients. The most common hypothesis test involves testing that null hypothesis of:
$$ H_{0} : \beta_{1} = 0 $$

versus

$$ H_{A} : \beta_{1} \ne 0 $$

This checks if $\beta_{1}$ has an effect on $Y$ or not.

### T-Statistic

$$ t = \dfrac{\hat{\beta}_{1} - \beta_{0}}{SE(\hat{\beta}_{1})} $$

* **T-statistic is like a pregame for p-value**
* We can get p-value for this statistic
* We need t-value of 2 for p-value of 5%.

### Residual Standard Error (RSE)

$$ RSE = \sqrt {\dfrac {1} {n - 2} RSS} = \sqrt{\dfrac {1} {n - 2} \sum_{i = 1} ^ {n} (y_{i} - \hat{y}_{i}) ^ 2}$$

where RSS is the *residual sum-of-squares* = $\sum_{i = 1} ^ {n} (y_{i} - \hat{y}_{i}) ^ 2 $

**RSE is RSS / degrees of freedom = rest of variance left unexplained.

*R-squared* or fraction of variance explained is:

$$ R ^ 2 = \dfrac{ TSS - RSS } { TSS } = 1 - \dfrac {RSS} {TSS} $$

where $ TSS = \sum_{i = 1} ^ {n} (y_{i} - \bar{y}) ^ 2 $ is the *total sum of squares*.

* **R-squared measures how much we reduced the TSS, how much we eliminated the variance**


$$ R ^ 2 = r ^ 2 $$

where 

$$ 
r = \dfrac {\sum_{i = 1} ^ {n} (x_{i} - \bar{x})(y_{i} - \bar{y}) } 
{ \sqrt {\sum_{i = 1} ^ {n} (x - \bar{x}) ^ 2} \sqrt {\sum_{i = 1} ^ {n} (y_{i} - \bar{y}) ^ 2}}
$$

* Just using TV advertisement, we reduced TSS of sales by 62%, which is huge.

## Multiple Linear Regression (10)

Model:

$$Y = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + ... + \beta_{n}x_{n} $$

* Multiple linear regression tries to fit a plane in a 3-dimensional scatter. Or n-1-dimensional figure in a n-dimensional scatter.
* The ideal scenario is when the predictors are uncorrelated
    * a balanced design:
        * Each coefficient can be estimated and tested separately.
        * A unit change in $X_{1}$  doesn't affect $X_{2}$.
    * correlations amongst predictors cause problems:
        * The variance of all coefficients tends to increase, sometimes dramatically
        * Interpretations become hazardous - when $X_{j}$ changes, everything else changes.
   * *Claims of causality* should be avoided.
   
### Estimation

$$ \hat{y} = \hat{\beta_{0}} + \hat{\beta}_{1}x_{1} + \hat{\beta}_{2}x_{2} + ... + \hat{\beta}_{p}x_{p} $$

* Just because TV, radio, newspaper independent of each other are adding sales (except newspaper), newspaper and radio have a .35 correlation (newspaper sales soaked up by radio), etc.

## Some Important Questions (11)

1. Is at least one of the predictors $X_{1}, X_{2}, ..., X_{p}$ useful in predicting the response?
2. Do all the predictors help to explain $Y$, or is only a subsset of the predictors useful?
3. How well does the model fit the data?
4. Given a set of predictor values, what response value should we predict, and how accurate is our prediction?

### Answers

1. We use F-statistic:

    $$ F = \dfrac {(TSS - RSS) / p} {RSS / (n - p -1)} \sim F_{p, n - p -1} $$
    * Will discuss how effect the predictors are. Bigger is better

2. The most direct approach is called *all subsets* or *best subsets*: we compute the least squares fit for all possible subsets.
    * However this is hard when you have lots of coefficients
    * We'll have an automated approach:
        * Forward selection:
            * Begin with *null model* - a model that contains an intercept but no predictions.
            * Fit p simple linear regs and add to null model
            * Add to that model the variable that results in lowest RSS
            * Continue adding until you reach good p-value
        * Backward selection:
            * Remove the variable that has the least significant (lowest t-statistic)
            * Keep removing until satisfactory p-value

## Extensions of Linear Model (12)

Given $100,000, if we spent 1/2 on TV and 1/2 on Radio, may improve sales more than just put money all in 1. 
* In marketing this is called synergy.
* How do we account for this?

### Make a new model

* $$f = 
\beta_{0} + 
\beta_{1} * x_{1} + 
\beta_{2} * x_{2} + 
\beta_{3} (x_{1} * x_{2}) + ... + \epsilon
$$

* You'll see a t-statistic of radio * TV = 21, which is significant
* R-squared goes from 89% to 97% when we add interaction
* The heirarchy principles:
    * If we include an interaction (radio x tv) we should also include the main effect (radio, tv)

### Interactions between qualitative and quantitative

* balance $\approx \beta_{0} + \beta_{1} * $ income $_{i}$ + $
\begin{cases}
\beta_{2} \text{ if ith person is a student} \\
0 \text{ if ith person is not a student}
\end{cases} $
* balanse $\approx \beta_{0} + \beta_{1} * $ income$_{i} + 
\begin{cases}
\beta_{2} + \beta_{3} * \text{income; if student} \\
0 \text{ if not student}
\end{cases} $
* You can also add squares, cubes, etc. as new variables

```r
fit2 = lm(medv~lstat + age, data = Boston)
#linear model of medv with respect to lstat and age
summary(fit2)
# we see age as significant but not as significant as lstat
```

```r
fit3 = lm(medv~. , Boston) # ~. uses all but medv as predictors
summary(fit3) #note that age is no longer significant in light of others
```

```r
par(mfrow = c(2, 2))
plot(fit3)
```

```r
fit4 = update(fit3, ~. -age -indus) 
#basically do lm again on fit3 data except age and indus
summary(fit4) #everything else is significant
```

```r
fit5 = lm(medv~lstat*age, Boston) #interactions of lstat, age
summary(fit5)
```

```r
fit6 = lm(medv~lstat +I(lstat^2), Boston) #protect lstat^2
summary(fit6)
```
