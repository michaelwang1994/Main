# Classification (14)

In machine learning, there is a big emphasis on classification (whether an email is spam or not, whether a patient will die or not).

* Qualitative variables take values in unordered set $C$.
    * eye color $\in$ { brown, blue, green }
    * email $\in$ {spam, ham}
* We are to take input $X$ and return on of the discrete values in $C$
    * We are more interested in estimating probabilities $X$ belongs to each category $C$.

Example using credict card default:

* Scatterplot of balance and income (balance is the factor that separates defaulters and non-defaulters)
    * A box plot of distributions of balance vs. income 
        * A box plot includes the median, the 75% quartile, the 25% quartile, the ranges, and the out-liers.
            * **Look at box plots and scatterplots of data!**

### Can we use linear regression?

Yes.

* Set **default** classification as:

$$ Y = \begin{cases}
0 \text{ if No} \\
1 \text{ if Yes}
\end{cases} $$

While we think that linear regression is perfect for the task, it often produces probabilities > 1 and < 0, so ** logistic regression** is better.
* i.e. with a scatter of 0s and 1s, the linear regression line can go under 0
* whereas logistic regression will cap and 0 and 1

Also, if you have 3 classifications:

$$ Y = \begin{cases}
1 \text{ if stroke} \\
2 \text{ if drug overdose} \\
3 \text{ if epileptic seisure}
\end{cases} $$

* linear regression will think stroke < drug overdose < epileptic seisure, which is not true. Logistic regression is the way to go!
    
    

# Logistic Regression and Maximum Likelihood (15)

Let's write $p(X) = \Pr(Y = 1| X)$ for short and consider using **balance** to predict **default**. Logistic regression uses the form:

$$ p(X) = \dfrac {e^{\beta_{0}+\beta_{1}X}}{1 + e^{\beta_{0}+\beta_{1}X}}$$

Where $e \approx 2.71828$. No matter what values the variables take, it will always be between 0 and 1.

Rearranging the formula gets us this:

$$ \log \big( \dfrac {p(X)}{1 - p(x)} \big) = \beta_{0} + \beta_{1}X$$

This is called the **log odds** or **logit** transformation of $p(X)$.

We use **maxixmum likelihood** to estimate the parameters.

$$ \ell(\beta_{0}, \beta) = \prod_{i:y_{i} = 1} p(x_{i}) \prod_{i:y_{i} = 0} (1 - p(x_{i})) $$

This **likelihood** gives the probability of the observed zeros and ones in the data. 
* This is the products of the probability of 0s and 1s, so choose the ones that give you the highest overall probability.
* In R use **glm**.
    * Will give us coefficients, std. error, z-statistic, p-value of the variables.
        * No one cares about the p-value for the intercept.

We can predict probabilities with this model for people with 1k in their balance:

$$ \hat{p}(X) = \dfrac {e^{\hat{\beta_{0}}+\hat{\beta_{1}}X}}{1 + e^{\hat{\beta_{0}}+\hat{\beta_{1}}X}} = 0.006 $$

# Multivariate Logistic Regression and Confounding (16)

$$ \log \dfrac{p(X) }{1 - p(X)} = \beta_{0} + \beta_{1}X_{1} + ... + \beta_{p}X_{p} $$

$$ p(X) = \dfrac {e^{\beta_{0} + \beta_{1}X_{1} + ... + \beta_{p}X_{p}}} {1 + e^{\beta_{0} + \beta_{1}X_{1} + ... + \beta_{p}X_{p}}} $$

From the previous example, we'll see that the coefficient for **student** is now negative where it as positive before. What happened?

* Students have higher balances than non-students, so their marginal default rate is higher than non-students. But for each level of student, they default less than non-students. **Multivariate logistic regression will reveal this**.

# Case-Control Sampling and Multiclass Logistic Regression (17)

In the South Africa data, there are 160 cases, 302 controls - 35% are cases. Yet the prevalence of MI in this region is 5%.

* With case-control samples, we can estimate the regression parameters $\beta_{j}$ accurately (if our model is correct); the constant term $\beta_{0}$ is incorrect.
* Let's correct the estimated intercept by a simple transformation:

$$ \hat{\beta}_{0}* = \hat{\beta_{0}} + \log \dfrac {\pi} {1- \pi} - \log \dfrac {\pi_{0}} {1 - \pi_{0}}$$

With $\pi = $ actual heart disease % and $\pi = $ experiment heart disease %. **This shifts $\beta_{0}$ down to the appropriate population level.**

### Logistic Regression with more than two classes...

$$ \Pr(Y = k|X) = \dfrac 
{e ^ {\beta_{0k} + \beta_{1k}X_{1} + ... + \beta_{pk}X_{p}}} 
{\sum_{\ell = 1}^{k} e ^ {\beta_{0 \ell} + \beta_{1 \ell}X_{1} + ... + \beta_{p \ell}X_{p}}}$$

Used in **glmnet** in R.

````r
pairs(Smarket, col = Smarket$Direction)
# scatterplots based on all combos of variables, 
# colors using market directions
````

````r
glm.fit = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume,
             data = Smarket, family = binomial)
summary(glm.fit)
# no significant coefficients here...
````

````r
glm.probs = predict(glm.fit, type = "response")
# make predictions on training data
glm.probs[1:5]
# all around 50%, not surprising...
````

````r
glm.pred = ifelse(glm.probs > 0.5, "Up", "Down")
# if glm.probs > .5, then call it "up", else call it "down
attach(Smarket)
````

````r
table(glm.pred, Direction)
# see there are lots of off predictions
````

````r
mean(glm.pred == Direction)
# you are right .52 of the time. A 2% improvement!
````

````r
# make training and test data
train = Year < 2005
glm.fit = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume,
             data = Smarket, family = binomial, subset = train)
# only use training data
glm.probs = predict(glm.fit, newdata = Smarket[!train, ], type = "response")
# apply it to the test dataset
glm.pred = ifelse(glm.probs > 0.5, "Up", "Down")
Direction.2005 = Smarket$Direction[!train]
table(glm.pred, Direction.2005)
````

````r
mean(glm.pred==Direction.2005)
# we are doing worse than we were! Overfitting?
````

````r
glm.fit = glm(Direction~Lag1+Lag2,
             data = Smarket, family = binomial, subset = train)
# just using lag 1, 2
glm.probs = predict(glm.fit, newdata = Smarket[!train, ], type = "response")
# apply it to the test dataset
glm.pred = ifelse(glm.probs > 0.5, "Up", "Down")
Direction.2005 = Smarket$Direction[!train]
mean(glm.pred==Direction.2005)
# we see an improvement, using less variables
````
