## What is Statistical Learning (03)

Given that Sales are based on TV, radio, and newspaper advertisement, let's make a model:

$$Sales \approx f (TV, Radio, Newspaper)$$

Set variables as a matrix:

$$X = \begin{pmatrix} X_{1} \\ X_{2} \\ X_{3} \end{pmatrix}$$

Now we write our model as:

$$Y = f(X) + \epsilon$$

Where $\epsilon$ is the error term.

With a good $f$ we can make predictions of $Y$ at new points $X = x$.
* As a result, we may see which components of $X$ are impactful and which are not.

Is there an ideal $f(X)$? In particular, what is a good value for $f(X)$ at any selected value of $X$, say $X = 4$? There can be many $Y$ values at $X = 4$, A good value is:

$$f(4) = E(Y|X = 4)$$

$E(Y|X = 4)$ means *expected value* (average) of $Y$ given $X = 4$. The ideal expected value results in the *regression function*.

### The regression function $f(x)$

* Is also defined for vector X; e.g. $f(x) = f(x_{1}, x_{2}, x_{3}) = E(Y| X_{1} = x_{1}, X_{2} = x_{2}, X_{3} = x_{3})$
* Is the *ideal* or *optimal* predictor of $Y$ with regard to mean-squared preddiction error: $f$ is the funciotn that minimizes $E[(Y - g(X))^2 | X = x]$ over all functions $g$ at all points $X = x$.

### Estimating $f$

You can calculate neighborhoods of X:

$\hat{f}(x) = Ave( Y | X \in \mathcal{N}(x))$ where $\mathcal{N}$ is a neighborhood of $x$

## Where nearest neighbor doesn't work so well (04)

The following is called "smoothing"
* Nearest neighbor methods tend to be far away in high dimensions (curse of dimensionality).
* You have to expand your neighborhoods drastically in order to capture data in high dimensions

### How do we deal with this?

Use a linear model (lasso):
$$f_{L}(X = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + ... + + \beta_{p}X_{p})$$

* HOWEVER, you NEVER want to make things 100% accurate, because that's called overfitting. You always want a "good fit"
* Parsimony vs. black-box
    * We often prefer a simpler model involving simpler methods, dispite inaccuracies
    
## Assessing Model Accuracy (05)

Split up the model in terms of *training* and *test* datasets.
* you want a model that **minimizes the distance between mean squared error of training and test data.**

### Bias-Variance Trade-off

$$E (y_{0} - \hat{f}(x_{0}))^2 = Var(\hat{f}(x_{0})) + [Bias(\hat{f}(x_{0}))]^2 + Var(\epsilon)$$
* Bias is MSE with training data based on # of variables
* Variance is MSE with test data based on # of variables
* Make the trade off when $Bias = Var$

## Classification Examples (06)

How do you classify spam vs. good emails (KNN)?

Is there an ideal classifying $C(X)$? Given $K$ elements,

$$ p_{k}(x) = Pr(Y = k | X = x), k = 1, 2, ..., K $$

These are the *conditional class probabilities* at $x$. Then the *Bayes optimal classifer* at $x$ is

$ C(X) = j $ if $ p_{j}(x) = max\{p_{1}(x),p_{2}(x), ... p_{K}(x)\} $

We measure the success of the classifying by:

$$ Err_{Te} = Ave_{i \in Te}I[y_{i} \ne \hat{C}(x_{i})] $$

**Bayes decision boundary** is the line in which classes are set. K-Nearest Neighbor will achieve a similar result (depending on the number of N nearest neighbors).
