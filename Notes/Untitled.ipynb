## Basic Simulation from Gaussian Distribution

Using a univariate distribution

```r
x = rep(1:1000)
y = rnorm(1000)
y_1 = rnorm(1000, sd = 1)
y_2 = rnorm(1000, sd = 1.1)

getReturns <- function (y) {
    y_new = rep(0, 1000)
    for (i in 1:1000) {
        y_new[i] = y[i] - y[i + 1]
    }
    return(y_new)
}

returns_0 = getReturns(y)
returns_1 = getReturns(y_1)
returns_2 = getReturns(y_2)
```

```r
plot(density(y), col = "red", ylim = c(0, .5))
lines(density(y_1), col = "blue")
lines(density(y_2), col = "green")
```

```r
plot(density(returns_0), col = "red")
lines(density(returns_1), col = "blue")
lines(density(returns_2), col = "green")
```
